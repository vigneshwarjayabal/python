```python
# Code to find the BMI based on user's weight and heigth.
# BMI = weight/(height*height)

h = int(input("enter your height:")) / 100
w = int(input("enter your weight"))


bmi = w / (h*h)

if bmi < 18.5:
  print(f"your BMI is {bmi} and your are UnderWeight")
elif bmi >=18.5 and bmi < 25:
  print(f"your BMI is {bmi} and your are Normal")
elif bmi >=25 and bmi < 30:
  print(f"your BMI is {bmi} and your are Slight OverWeight")
elif bmi >= 30 and bmi < 35:
  print(f"your BMI is {bmi} and your are Obese")
else:
  print(f"your BMI is {bmi} and your are clinicall Obese")


```
