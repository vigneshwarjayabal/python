```python
# You can specify that a function can have ONLY positional arguments, or ONLY keyword arguments.
# To specify that a function can have only positional arguments, add , / after the arguments.

def add(a,b,c,/):
  sum = a + b + c
  return sum

add(10,20,30)
```

```python
add(10,20,c=100)
```

```python
# Without the , / you are actually allowed to use keyword arguments even if the function expects positional arguments
# But when adding the , / you will get an error if you try to send a keyword argument.

# To specify that a function can have only keyword arguments, add *, before the arguments.
#Without the *, you are allowed to use positionale arguments even if the function expects keyword arguments
# But with the *, you will get an error if you try to send a positional argument.

def add(*,a,b,c):
  sum = a + b + c
  return sum

add(a=100,b=200,c=10)
```

```python
add(100,b=20,c=100)
```

```python
#You can combine the two argument types in the same function.
#Any argument before the / , are positional-only, and any argument after the *, are keyword-only.

def my_function(a, b, /, *, c, d):
  print(a + b + c + d)

my_function(5, 6, c = 7, d = 8)
```

#**Recursion in python:**

image.png



```python
# example 1
'''The factorial function:
The factorial of a non-negative integer (n) = the product of all positive integer less than or equal to n
It is often denoted as n!
0! = 1 base case
n! = (n * (n-1))! recursive case '''

def factorial(n): #1
  if n == 0: #2
    return 1 #3
  else: #4
    return n * factorial(n-1) # 5

''' line 2 and 3 are base case and line 5 is recursive case'''
print(factorial(6))
```

```python
# example 2
''' The fibonacci sequence
It is a another classical example for recursion.
each number in the sequence is the sum of the two prceeding ones,it usually starting with 0 and 1
fib(0) = 0
fib(1) = 1 this two is base case
fib(n) = fib(n-1)+fib(n-2)for n>1 this is recursive case '''

def fib(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fib(n-1) + fib(n-2)

print(fib(6))
```

```python
''' sample code for how to get input for dictionary from the user '''

n = int(input())
student = {}
for _ in range(n):
  line = input().split()
  name = line[0]
  marks = list(map(float,line[1:]))
  student[name] = marks

print(student)
```

```python
''' sample code for how to get input from the user and store in nested list '''
n = int(input())
student = []
for _ in range(n):
  name = input()
  score = float(input())
  student.append([name,score])
print(student)
```

```python
''' Map Function :
  Syntax:
  map(function,iterable)
  fucntion is that you want to apply to each item of the iterable, it can be built-in ,lambda function or any other callable object that accepts
  one or more arugment.
  iterable : list,tuple,dict,set'''
```

```python
def square(x):
  return x**2
number = [1,2,3,4,5,6,7]
squars = map(square,number)
print(list(squars))
```

```python
number = [2,4,6,8,10,12,14]
squares = map(lambda x: x**2, number)
print(list(squares))
```
