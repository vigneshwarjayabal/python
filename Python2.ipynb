```python
# example 1
'''The factorial function:
The factorial of a non-negative integer (n) = the product of all positive integer less than or equal to n
It is often denoted as n!
0! = 1 base case
n! = (n * (n-1))! recursive case '''

def factorial(n): #1
  if n == 0: #2
    return 1 #3
  else: #4
    return n * factorial(n-1) # 5

''' line 2 and 3 are base case and line 5 is recursive case'''
print(factorial(6))
```

```python
# example 2
''' The fibonacci sequence
It is a another classical example for recursion.
each number in the sequence is the sum of the two prceeding ones,it usually starting with 0 and 1
fib(0) = 0
fib(1) = 1 this two is base case
fib(n) = fib(n-1)+fib(n-2)for n>1 this is recursive case '''

def fib(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fib(n-1) + fib(n-2)

print(fib(6))
```

```python
# Using __init__() with Methods
# The __init__() method can also work in conjunction with other methods to make the class functional.

class BankAccount:
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} deposited. New balance: {self.balance}")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"{amount} withdrawn. Remaining balance: {self.balance}")
        else:
            print("Insufficient balance!")

# Creating an object
account = BankAccount("Vigneshwar", 1000)

# Using methods
account.deposit(500)  # Output: 500 deposited. New balance: 1500
account.withdraw(200) # Output: 200 withdrawn. Remaining balance: 1300
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

rect = Rectangle(5, 10)
print("Area:", rect.area())  # Output: Area: 50
print("Perimeter:", rect.perimeter())  # Output: Perimeter: 30
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Invalid amount")

    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
print("Balance:", account.get_balance())  # Output: Balance: 1500

# Trying to access private attribute directly will result in an error:
# print(account.__balance)  # AttributeError
```

```python
class Animal:
    def sound(self):
        return "Some sound"

class Dog(Animal):
    def sound(self):
        return "Bark"

class Cat(Animal):
    def sound(self):
        return "Meow"

animals = [Dog(), Cat(), Animal()]

for animal in animals:
    print(animal.sound())
```

```python
class Dog:
    def speak(self):
        return "Bark"

class Cat:
    def speak(self):
        return "Meow"

def make_sound(animal):
    print(animal.speak())

dog = Dog()
cat = Cat()

make_sound(dog)  # Output: Bark
make_sound(cat)  # Output: Meow
```
